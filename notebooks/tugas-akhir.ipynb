# Membaca PDF

```python
# Membuka dokumen PDF
pdf_file_path = "/content/drive/MyDrive/Tugas Akhir heru/Juknis TIF 2024-2025_Fix.pdf"
pdf_file = open(pdf_file_path, "rb")

# Membuat objek pembaca PDF
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Mendapatkan jumlah halaman dalam dokumen PDF
num_pages = len(pdf_reader.pages)

# Membaca teks dari setiap halaman
for page_num in range(num_pages):
    page = pdf_reader.pages[page_num]
    text = page.extract_text()
    print(f"Halaman {page_num + 1}:\n{text}\n")

# Menutup file PDF
pdf_file.close()
```

# Membuat Model

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
```

```python
# 1. Load PDF
loader = PyPDFLoader("/content/drive/MyDrive/Tugas Akhir heru/Juknis TIF 2024-2025_Fix.pdf")
pages = loader.load()
```

```python
# 2. Split text
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50,
    separators=["\n\n", "\n", " ", ""]
)
texts = text_splitter.split_documents(pages)
```

```python
# 3. Create embeddings
embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/multi-qa-mpnet-base-dot-v1"
)
```

```python
# 4. Create vector store
vector_store = FAISS.from_documents(texts, embeddings)
```

```python
# 5. Save vector store
vector_store.save_local("faiss_index")
print("Vector store saved successfully!")
```

# Setup Model RAG

```python
from transformers import AutoTokenizer,AutoModelForSeq2SeqLM, pipeline
from langchain.llms import HuggingFacePipeline
from langchain.chains import RetrievalQA
import torch
```

```python
from huggingface_hub import login
login()
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# 1. Load model dan tokenizer (Gunakan AutoModelForSeq2SeqLM)
model_name = "google/flan-t5-small"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)  # Pastikan menggunakan Seq2Seq model
```

```python
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,  # Gunakan float16 untuk lebih hemat memori
    device_map="auto"  # Pindahkan ke GPU jika tersedia
)
```

```python
# 2. Create pipeline
pipe = pipeline(
    "text2text-generation",
    model=model,
    tokenizer=tokenizer,
    max_length=512,
    temperature=0.3
)
```

```python
# 3. Create QA chain
qa = RetrievalQA.from_chain_type(
    llm=HuggingFacePipeline(pipeline=pipe),
    chain_type="stuff",
    retriever=vector_store.as_retriever(),
    return_source_documents=True
)
```

```python
# Test model
query = "sebutkan syarat skripsi?"
result = qa.invoke({"query": query})
print("Answer:", result["result"])
```
